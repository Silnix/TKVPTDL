# Vẽ biểu đồ Line, Line + point

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

# Dữ liệu mẫu về thu nhập bình quân/năm của một người trong 10 năm
nam = [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]
thu_nhap = [250000000, 300000000, 230000000, 450000000, 460000000, 600000000, 400000000, 450000000, 480000000, 600000000]

# Vẽ biểu đồ dạng line + point, marker='o': Các điểm dữ liệu trên biểu đồ sẽ có dạng hình tròn, linestyle='-': Dùng để chỉ định kiểu đường thẳng, color='red': Màu của đường là màu đỏ (có thể viết tắt: g, r, b)
plt.plot(nam, thu_nhap, marker='o', linestyle='-', color='green')
# Thêm tiêu đề và nhãn trục
plt.title('Biểu đồ thu nhập cá nhân trong 10 năm')
plt.xlabel('Năm')
plt.ylabel('Thu nhập (VNĐ)')

# Hiển thị lưới trên biểu đồ
plt.grid(True)

# Hiển thị đủ nhãn từng năm trên trục hoành, nếu cần
#plt.xticks(nam)  # Đặt các giá trị nhãn trên trục x là các năm
plt.xticks(nam, rotation=45)  # Xoay nhãn trên trục hoành 45 độ

# Định dạng nhãn trục y để không bị ở dạng số khoa học
def format_currency(x, pos):
    #return f'{int(x):,} VNĐ'
    return f'{int(x):,}'
    #return int(x)

formatter = FuncFormatter(format_currency)
plt.gca().yaxis.set_major_formatter(formatter)
#plt.yticks(thu_nhap, rotation=45)
# Hiển thị biểu đồ
plt.show()
```

# Biểu đồ Scatter

```python
import numpy as np
import matplotlib.pyplot as plt
height = np.array([167,170,149,165,155,180,166,146,159,185,145,168,172,181,169])
weight = np.array([86,74,66,78,68,79,90,73,70,88,66,84,67,84,77])

colors = np.random.rand(15)
area = (30 * np.random.rand(15))**2

plt.xlim(140,200) #xác định hệ số chia trên trục hoành
plt.ylim(60,100) #xác định hệ số chia trên trục tung
plt.scatter(height,weight,s=area,c=colors)
plt.title("Chiều cao và cân nặng")
plt.xlabel("Chiều cao - cm")
plt.ylabel("Cân nặng - kg")

plt.show()
```

```python
#thêm nhãn số liệu cho mỗi cột
x_values = [0, 1, 2]
y_values = [10, 20, 50]
data_labels = ['10', '20', '50']
plt.bar(x = data_labels, height = y_values, color = 'green')

for i in range(len(data_labels)): # your number of bars
  plt.text(x = x_values[i], #takes your x values as horizontal positioning argument
  y = y_values[i]+1, #takes your y values as vertical positioning argument
  s = data_labels[i], # the labels you want to add to the data
  size = 9)
```

```python
# vẽ biểu đồ 2 biến
import numpy as np

men_means, men_std = (20, 35, 30, 35, 27), (2, 3, 4, 1, 2)
women_means, women_std = (25, 32, 34, 20, 25), (3, 5, 2, 3, 3)

ind = np.arange(len(men_means))  # the x locations for the groups
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(ind - width/2, men_means, width, yerr=men_std,
                label='Men')
rects2 = ax.bar(ind + width/2, women_means, width, yerr=women_std,
                label='Women')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(ind)
ax.set_xticklabels(('G1', 'G2', 'G3', 'G4', 'G5'))
ax.legend()

plt.show()
```

# Biểu đồ Pie

```python
import numpy as np
plt.pie(x = np.array([10, 20, 50]), # giá trị của các nhóm
        labels = ['Nhóm A', 'Nhóm B', 'Nhóm C'], # Nhãn của các nhóm
        colors = ['red', 'blue', 'green'], # Màu sắc của các nhóm
        autopct = '%1.1f%%', # Format hiển thị giá trị %
        shadow = True
       )
plt.title('Biểu đồ tròn tỷ lệ % của các nhóm')
```

# Biểu đồ boxplot

```python
import numpy as np
x = np.random.randn(100) + np.arange(0, 100) * 0.5
y = np.random.randn(100) + np.arange(0, 100) * 1.0 + 10
z = np.random.randn(100) + np.arange(0, 100) * 2 - 15

#plt.boxplot([x, y, z], labels = ['x', 'y', 'z'], showfliers = True)
plt.boxplot([x, y, z], labels = ['x', 'y', 'z'])
plt.title('Biểu đồ Boxplot')
plt.xlabel('Classes')
plt.ylabel('Gía trị của x, y, z')
```

# Biểu đồ Histogram

```python
import matplotlib.pyplot as plt

# Dữ liệu điểm số của sinh viên
scores = [65, 70, 72, 75, 78, 80, 82, 85, 88, 90, 92, 95, 97, 100]

# Tạo histogram, tạo các khoảng điểm thông qua bins,
plt.hist(scores, bins=[60, 70, 80, 90, 100], edgecolor='black')

# Thêm tiêu đề và nhãn trục
plt.title("Biểu đồ Histogram về điểm số sinh viên")
plt.xlabel("Điểm số")
plt.ylabel("Số lượng sinh viên")

# Hiển thị biểu đồ
plt.show()
```
